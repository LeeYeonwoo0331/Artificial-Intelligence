```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread("hist_half.jpg" ,cv.IMREAD_GRAYSCALE)
```

```python
hist_origin = cv.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist_origin)
plt.show()
```

```python
img_float = np.float32(img)
min_val, max_val, min_loc, max_loc = cv.minMaxLoc(img_float)
img_cal = (img_float - min_val) / (max_val - min_val) * 255
img_cal = np.uint8(img_cal)
```

```python
hist_norm = cv.calcHist([img_cal], [0], None, [256], [0, 256])
plt.plot(hist_norm)
plt.show()
```

```python
img_stack = np.hstack((img, img_cal))
```

```python
cv.imshow("img_stack", img_stack)
cv.waitKey(0)
cv.destroyAllWindows()
```
